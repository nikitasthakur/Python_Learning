```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
pollutant_data = pd.read_csv("./datasets/UFOPOLLUTANTS.csv")
pollutant_data.head()
```

```python
pollutant_data = pollutant_data[:1000]
```

```python
sns.axes_style()
```

```python
sns.distplot(pollutant_data["NO2.AQI"])
```

```python
sns.set_style("whitegrid",{
                           "axes.facecolor": "white",
                           "axes.edgecolor": "black",
                           "axes.labelcolor": "green",
                           "grid.color": "#000000",
                           "font.family":"fantasy",
                           "grid.linestyle": "-.",
                           "axes.spines.top": False,
                           "axes.spines.right": False,})

sns.distplot(pollutant_data["NO2.AQI"] , 
                            color = "maroon")
```
