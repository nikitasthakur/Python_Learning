```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
car_data = pd.read_csv("./datasets/mpg.csv")
car_data.head()
```

```python
plt.subplots(figsize = (12,6))

sns.boxplot(x = "cylinders",
            y = "horsepower",
            data = car_data)
```

```python
plt.subplots(figsize = (12,6))

sns.violinplot(x = "cylinders",
            y = "horsepower",
            data = car_data)
```

```python
plt.subplots(figsize = (12,6))

sns.violinplot(x = "cylinders",
            y = "horsepower",
            data = car_data,
            inner = None)

sns.swarmplot("cylinders",
              "horsepower",
              data = car_data,
              color = "black",
              size = 3)
```

```python
sns.set(style = "whitegrid",
        color_codes = True)
```

```python
plt.subplots(figsize = (12,6))

sns.stripplot("cylinders",
              "horsepower",
              data = car_data,
              size = 6)
```

```python
plt.subplots(figsize = (12,6))

sns.stripplot("cylinders",
              "horsepower",
              data = car_data,
              size = 6,
              jitter = True)
```

```python
plt.subplots(figsize = (12,6))

sns.swarmplot("cylinders",
              "horsepower",
              data = car_data)
```

```python
plt.subplots(figsize = (12,6))

sns.swarmplot("cylinders",
              "horsepower",
              data = car_data,
              hue = "origin")
```

```python
plt.subplots(figsize = (12,6))

sns.barplot("cylinders",
              "horsepower",
              data = car_data,
              hue = "origin")
```

```python
plt.subplots(figsize = (12,6))

sns.countplot("origin",
              data = car_data,
              palette = "Accent")
```

```python
plt.subplots(figsize = (12,6))

sns.pointplot("cylinders",
              "horsepower",
              data = car_data)
```

```python
sns.catplot(x = "cylinders",
            y = "horsepower",
            data = car_data,
            row = "origin",
            kind = "box",
            height = 8,
            aspect = 2)
```
