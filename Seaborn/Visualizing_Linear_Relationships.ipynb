```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
pollution_data = pd.read_csv("./datasets/UFOPOLLUTANTS.csv")
pollution_data = pollution_data[:1001] 
```

```python
sns.lmplot(x = "CO.1st.Max.Value",
           y = "NO2.1st.Max.Value",
           data = pollution_data)
```

```python
sns.lmplot(x = "O3.1st.Max.Hour",
           y = "O3.AQI",
           data = pollution_data)
```

```python
sns.lmplot(x = "O3.1st.Max.Hour",
           y = "O3.AQI",
           x_jitter = .4,
           y_jitter = 0.1,
           data = pollution_data)
```

```python
sns.lmplot(x = "O3.1st.Max.Hour",
           y = "O3.AQI",
           data = pollution_data,
           height = 5,
           aspect = 2.5)
```

```python
sns.lmplot(x = "O3.1st.Max.Hour",
           y = "O3.AQI",
           data = pollution_data,
           x_estimator = np.mean)
```

```python
sns.lmplot(x = "O3.1st.Max.Hour",
           y = "O3.AQI",
           data = pollution_data,
           row = "year")
```

```python
sns.lmplot(x = "O3.1st.Max.Hour",
           y = "O3.AQI",
           data = pollution_data,
           col = "year")
```
