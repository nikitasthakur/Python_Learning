```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pollution_data = pd.read_csv("./datasets/UFOPOLLUTANTS.csv")
pollution_data = pollution_data.head(1000)
```

```python
sns.distplot(pollution_data['NO2.AQI'])
```

```python
sns.distplot(pollution_data['CO.AQI'])
```

```python
dist = sns.distplot(pollution_data['NO2.AQI'], color = "Purple")

dist.set_title("POLLUTION STATS")

dist.set_xlabel("Air Quality index of NO2")

dist.set_ylabel("Density")
```

```python
dist = sns.distplot(pollution_data['NO2.AQI'], 
                    color = "Purple",
                    vertical = True)

dist.set_title("POLLUTION STATS")

dist.set_ylabel("Air Quality index of NO2")

dist.set_xlabel("Density")
```

```python
dist = sns.distplot(pollution_data['NO2.AQI'],
                    
                    kde_kws = {"color":"Blue",
                               "lw" : 2,
                               "label": "KDE"},
                    
                    hist_kws = {"histtype" : "step",
                                "linewidth" : 2,
                                "alpha" : 0.7,
                                "color" : "Green",
                                "label" :  "hist"})


dist.set_title("POLLUTION STATS")

dist.set_xlabel("Air Quality index of NO2")

dist.set_ylabel("Density")
```

```python
dist = sns.distplot(pollution_data['NO2.AQI'], 
                    color = "Purple",
                    kde = False)

dist.set_title("POLLUTION STATS")

dist.set_xlabel("Air Quality index of NO2")

dist.set_ylabel("Density")
```

```python
dist = sns.distplot(pollution_data['NO2.AQI'],
                    bins = 40,
                    kde = False)

dist.set_title("POLLUTION STATS")

dist.set_xlabel("Air Quality index of NO2")

dist.set_ylabel("Density")
```

```python
dist = sns.distplot(pollution_data['NO2.AQI'], 
                    color = "Purple",
                    hist = False)

dist.set_title("POLLUTION STATS")

dist.set_xlabel("Air Quality index of NO2")

dist.set_ylabel("Density")
```

```python
rug = sns.rugplot(pollution_data['NO2.AQI'], 
                   height = 0.7,
                   axis = "x")

rug.set_title("POLLUTION STATS")

rug.set_xlabel("Air Quality index of NO2")

rug.set_ylabel("Density")
```

```python
plt.subplots(figsize = (20,5))

dist = sns.distplot(pollution_data['NO2.AQI'], 
                    bins = 95,
                    rug = True)

dist.set_title("Histogram, KDE and Rugplot")

dist.set_xlabel("Air Quality index of NO2")

dist.set_ylabel("Density")
```

```python
dist = sns.distplot(pollution_data['NO2.AQI'], 
                    hist = False,
                    rug = True)

dist.set_title("KDE and Rugplot")

dist.set_xlabel("Air Quality index of NO2")

dist.set_ylabel("Density")
```

```python
sns.set(color_codes = True)

plot = sns.kdeplot(pollution_data['NO2.AQI'], 
                    shade = True,
                    color = "b",
                    legend = True)

plot.set_title("KDE and Rugplot")

plot.set_xlabel("Air Quality index of NO2")

plot.set_ylabel("Density")
```

```python
with sns.axes_style("ticks"):
    
    sns.jointplot("NO2.Mean",
                  "CO.Mean",
                  data = pollution_data,
                  space = 1,
                  color = "g")
```

```python
sns.kdeplot(pollution_data["NO2.Mean"],
            pollution_data["CO.Mean"],
            cmap = "PRGn",
            cbar = True)
```

```python
sns.kdeplot(pollution_data["NO2.Mean"],
            pollution_data["CO.Mean"],
            clip = (-10,60))
```

```python
sns.kdeplot(pollution_data["NO2.Mean"],
            pollution_data["CO.Mean"])

sns.rugplot(pollution_data["NO2.Mean"],
            color = "r",
            height = .05)

sns.rugplot(pollution_data["CO.Mean"],
            color = "b",
            vertical = True,
            height = .05)
```

```python
sns.pairplot(pollution_data,
             height = 5,
             vars = ["NO2.Mean",
                     "CO.Mean",
                     "SO2.Mean",
                     "O3.Mean"])
```

```python
pollutants_mean = pollution_data[['NO2.Mean',
                                  'CO.Mean',
                                  'SO2.Mean',
                                  'O3.Mean']]
```

```python
corrmatrix = pollutants_mean.corr()

plt.subplots(figsize = (7,7))

sns.heatmap(corrmatrix,
            vmax =.8,
            square = True,
            annot = True,
            fmt = ".2f",
            cmap = "YlGnBu")
```
