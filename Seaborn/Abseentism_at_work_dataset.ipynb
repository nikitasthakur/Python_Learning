```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
employee_data = pd.read_csv("./datasets/Absenteeism_at_work.csv")
employee_data.head()
```

```python
employee_data.describe()
```

```python
hist = sns.distplot(employee_data['Reason for absence'])
```

```python
plt.subplots(figsize = (12,6))
sns.distplot(employee_data['Reason for absence'],
             bins = 29,
             kde = False,
             rug = True)
```

```python
plt.subplots(figsize = (12,6))
sns.distplot(employee_data['Month of absence'],
             bins = 12,
             kde = False,
             rug = True)
```

```python
sns.jointplot(x = "Day of the week",
              y = "Reason for absence",
              data = employee_data,
              height = 6)
```

```python
sns.jointplot(x = "Seasons",
              y = "Reason for absence",
              data = employee_data,
              height = 6)
```

```python
pairplot = sns.pairplot(employee_data,
                        height = 4,
                        aspect = 1,
                        
                        x_vars = ["Distance from Residence to Work",
                                  "Body mass index",
                                  "Age" ],
                        
                        y_vars = ["Absenteeism time in hours", 
                                  "Work load Average/day "],
                        
                        kind = "reg",
                        hue = "Disciplinary failure")

#pairplot.fig.subplots_adjust(wspace = 0.5 , hspace = 0.1)
```
