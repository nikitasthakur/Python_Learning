```python
cars_data = pd.read_csv('./datasets/ZoomCars.csv')
```

```python
cars_data = cars_data.drop(columns=['id','url','page','fetched','price_currency','ad_id','location','description'])
```

```python
cars_data = cars_data.dropna()
```

```python
ChangeCase = lambda x:x.upper()
```

```python
cars_data['import'] = cars_data['current_location'].map({'In Tanzania' : 'No',
                                                         'Available For Import' :  'Yes'})
```

```python
cars_data['import'].unique()
```

```python
def drop_comma(x):
    return str(x).replace(',','')
```

```python
cars_data['price_value'].head(10)
```

```python
cars_data['price_value'] = cars_data['price_value'].apply(drop_comma)
```

```python
cars_data['price_value'].head(10)
```

```python
cars_data['price_value'] =  pd.to_numeric(cars_data['price_value'])
```

```python
cars_data['price_value']
```

```python
def category(price):
    price = int(price)
    
    if price <= 10000000:
        return 'cheap'
    elif price > 10000000 and price <= 30000000:
        return 'mid-range'
    elif price > 30000000 and price <= 70000000:
        return 'expensive'
    else:
        return 'luxury'
```

```python
cars_data.price_value.apply(category).head(10)
```

```python
cars_data['price'] = cars_data.price_value.apply(category).head(10)
```
