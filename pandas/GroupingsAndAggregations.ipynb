```python
array = objects_data.values.T.tolist()
```

```python
array
```

```python
array_index = array[1:3]
array_index
```

```python
index = pd.MultiIndex.from_arrays(array_index,
                                  names =['shape','colour'])
```

```python
multi_data = pd.DataFrame({'object' :array[0],
                          'quantity':array[3]},
                          index = index)
```

```python
multi_data
```

```python
multi_data.groupby(level=['shape']).sum()
```

```python
multi_data.groupby(level =['colour']).sum()
```

```python
multi_data.groupby(level=['shape','colour']).sum()
```

```python
multi_data.groupby(level=['shape','colour']).get_group(('rectangle','white'))
```

```python
multi_data.groupby([pd.Grouper(level='shape'),'object']).sum()
```

```python
multi_data.groupby([pd.Grouper(level=0),'object']).sum()
```

```python
multi_data.groupby(level=['shape','colour']).size()
```

```python
multi_data.groupby(level=['shape','colour']).describe()
```

```python
multi_data.groupby(level=['shape','colour']).aggregate(np.sum)
```

```python
multi_data.groupby(level=['shape','colour']).apply(np.sum)
```

```python
 multi_data.groupby(level=['shape'],as_index=False).agg(np.sum)
```

```python
multi_data.groupby(level=['shape','colour']).agg([np.sum,np.min,np.max,np.mean])
```

```python
multi_data.groupby(level=['shape']).transform(lambda x:x.mean())
```

```python
multi_data['average'] = multi_data.groupby(level=['shape']).transform(lambda x:x.mean())
```
